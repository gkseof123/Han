```python
import random


class BaseBall:
#인스턴스 변수 선언은 __init__()내부에....
#클래스로부터 생성된 인스턴스들이 공유하는 변수(전역변수, 공유변수)선언 클래스변수 
    player = 0
    outCount = 0
    strike = 0
    ball = 0
    
    @classmethod
    def getStatus(cls):
        cls.player += 1
        print('*******', cls.player, '번째 선수 출격*******')
        while(cls.outCount < 3):
            BaseBall.isStrike()
            if(cls.strike==3):
                cls.outCount+=1
                cls.player+=1
                print('선수교체')
                print('*******',cls.player,'번째 선수 출격*******')
                cls.strike = 0
                cls.ball = 0
            elif(cls.ball == 4):
                print('1루 출격')
                cls.player += 1
                print('******',cls.player,'번째 선수 출격******')
                cls.strike = 0
                cls.ball = 0
        print('쓰리아웃! 공수교체')
    @classmethod
    def isStrike(self):
        if(random.randint(0,1)==0):
            self.strike+=1
            print('공 던짐=>스트라이크!!')
            print(self.outCount,'아웃,',self.strike,'스트라이크,',self.ball,'볼')
            b=False
        else:
            self.ball+=1
            print('공 던짐=>볼!!')
            print(self.outCount,'아웃,',self.strike,'스트라이크,',self.ball,'볼')
            b=True
        return b


baseball= BaseBall()
baseball.getStatus()


           







        
        
```

```python
class AbnormalValueError(Exception):
    def __init__(self,message):
        self.message = message
        self.__tall = 160.5

    def getTall(self):
        return self.__tall

middle_talls = [ 149,155,130,165,161,170,152,163,193,168]

def tall_check(tall):
    if tall<140:
        raise AbnormalValueError("정상 키 범위보다 작습니다.")
    elif tall > 180 :
        raise AbnormalValueError("정상 키 범위보다 큽니다.")
for i, tall in enumerate(middle_talls):
    try:
        tall_check(tall)
    except AbnormalValueError as e :
        print(e.message)
        middle_talls[i] = e.getTall()

print(middle_talls)
print(sum(middle_talls)/len(middle_talls))
```

```python
instr = ""
upper, lower, digit, hangle, etc = [0] * 5
instr = input("문자를 입력하세여 :")

for x in instr :
    if ord('A') <= ord(x) and ord(x) <= ord('Z') :
        upper += 1
    elif ord('a') <= ord(x) and ord(x) <= ord('z') :
        lower += 1
    elif ord('가') <= ord(x) and ord(x) <= ord('힣') :
        hangle += 1
    elif ord('1') <= ord(x) and ord(x) <= ord('9') :
        digit += 1
    else :
         etc += 1
print("대문자 : %d 소문자 : %d 숫자 : %d 한글 : %d 기타 : %d" %(upper, lower, digit, hangle, etc) )
```
